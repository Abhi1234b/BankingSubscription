```python
l=['housing', 'loan', 'contact', 'duration', 'campaign', 'pdays',
       'previous', 'cons.price.idx', 'cons.conf.idx', 'euribor3m',
       'nr.employed','age1630', 'age3044', 'age4558', 'age5898', 'admin',
       'blue-collar', 'technician', 'services', 'retired', 'entrepreneur',
       'self-employed', 'housemaid', 'unemployed', 'married', 'single',
       'university.degree', 'high.school', 'basic.9y', 'professional.course',
       'basic']
X=df[l]
Y=df['y']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_sc=sc.fit_transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,roc_auc_score,classification_report
print(accuracy_score(y_train,lr.predict(X_train)))
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
```

```python
roc_auc_score(y_test,lr.predict_proba(X_test)[:, 1])
```

```python
roc_auc_score(y_train,lr.predict_proba(X_train)[:, 1])
```

```python
lr=LogisticRegression()
lr.fit(X_train_sc,y_train)
pred1=lr.predict(X_test_sc)
```

```python
print(accuracy_score(y_train,lr.predict(X_train_sc)))
print(accuracy_score(y_test,pred1))
print(classification_report(y_test,pred1))
```

```python
roc_auc_score(y_test,lr.predict_proba(X_test_sc)[:, 1])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
knn = KNeighborsClassifier()

knn.fit(X_train1,y_train1)
knn.score(X_test,y_test)
```

```python
svm = SVC()
svm.fit(X_train1,y_train1)
svm.score(X_test,y_test)
```

```python
predict_val1 = knn.predict(X_test_val)
predict_val2 = svm.predict(X_test_val)
```

```python
predict_val = np.column_stack((predict_val1,predict_val2))
predict_val
```

```python
predict_test1 = knn.predict(X_test)
predict_test2 = svm.predict(X_test)
```

```python
predict_test = np.column_stack((predict_test1,predict_test2))
predict_test
```

```python
rand_clf = RandomForestClassifier()

rand_clf.fit(predict_val,y_test_val)
```

```python
rand_clf.score(predict_test,y_test)
```
